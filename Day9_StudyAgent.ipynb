```python
import os
import getpass

os.environ["GEMINI_API_KEY"] = getpass.getpass("Enter Gemini API Key: ")
```

```python
from langchain_core.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun

search = DuckDuckGoSearchRun()

@tool
def web_search(query: str) -> str:
    """
    Search the web for recent or real-time information.

    Use this tool when:
    - latest news is needed
    - hiring updates are needed
    - current events are required

    If no results are found,
    return NO_RESULTS_FOUND.
    """

    try:
        result = search.run(query)

        if not result or len(result.strip()) == 0:
            return "NO_RESULTS_FOUND"

        return result[:1200]

    except Exception:
        return "NO_RESULTS_FOUND"
```

```python
print(web_search.invoke({
    "query": "TCS hiring 2026"
}))
```

```python
from langchain.agents import create_agent
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)

agent = create_agent(
    model=llm,
    tools=[web_search]
)

print("Agent created successfully.")
```

```python
result = agent.invoke({
    "messages": [
        {
            "role": "user",
            "content": "What is TCS's 2026 hiring quota?"
        }
    ]
})
```

```python
for i, m in enumerate(result["messages"]):

    print(f"\n[{i}] {type(m).__name__}")

    if hasattr(m, "content"):
        print("Content:")
        print(str(m.content)[:500])

    if hasattr(m, "tool_calls") and m.tool_calls:
        print("Tool Calls:")
        print(m.tool_calls)
```

```python
bad_result = agent.invoke({
    "messages": [
        {
            "role": "user",
            "content": "Search this URL and explain it: https://fake-domain-12345.com"
        }
    ]
})
```

```python
for i, m in enumerate(bad_result["messages"]):

    print(f"\n[{i}] {type(m).__name__}")

    if hasattr(m, "content"):
        print(str(m.content)[:500])
```
